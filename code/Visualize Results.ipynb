```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Saved Data

```python
transformer_data = pd.read_csv("../results/SSL_Effect_Transformer.csv")
transformer_data.head()
```

```python
transformer_data.columns
```

### Effect of SSL (Transformer)

```python
plt.figure(figsize=(8,6))
plt.title("Transformer Training Loss")

plt.plot(transformer_data['No SSL train'], label="No SSL")
plt.plot(transformer_data['seq train'], label="Only Latent")
plt.plot(transformer_data['add train'], label="Latent + Input")

plt.xlabel("Epoch Number")
plt.ylabel("MSE Loss")
plt.legend()
```

```python
plt.figure(figsize=(5,3))
plt.title("Transformer Training Loss")

plt.plot(transformer_data['No SSL train'], label="No SSL")
plt.plot(transformer_data['add train'], label="Latent + Input")

plt.xlabel("Epoch Number")
plt.ylabel("MSE Loss")
plt.legend()
```

### Effect of Transformer Imputation

```python
plt.figure(figsize=(8,6))
plt.title("Transformer Training Loss")

plt.plot(transformer_data['interp train'], label="Interpolation Imputation")
plt.plot(transformer_data['add train'], label="Transformer Imputation")

plt.xlabel("Epoch Number")
plt.ylabel("MSE Loss")
plt.legend()
```

### Without Missing Data

```python
plt.figure(figsize=(8,6))
plt.title("Transformer Training Loss with No Missing Data")

plt.plot(transformer_data['No SSL No mask train'], label="No SSL")
plt.plot(transformer_data['add  No mask train'], label="Latent + Input")

plt.xlabel("Epoch Number")
plt.ylabel("MSE Loss")
plt.legend()
```
